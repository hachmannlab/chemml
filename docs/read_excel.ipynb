## Read a custom excel sheet

This is a tutorial on how to read an excel sheet using the GUI. We use a random excel sheet to assist with the tutorial. The excel sheet consists of single column of SMILES codes of a few molecules. We read the SMILES codes, generate their Morgan Fingerprints, which are available through the RDKit library, and save them. 

The aim of the tutorial is to aid the user in reading an excel sheet containing data on which calculations are to be performed. 

A dedicated template for the same exists under the heading "_Entering Custom Dataset from Excel:_", in the _Template workflow_ tile of the GUI. For details regarding the file formats, inputs and outputs, we recommend that users view each of the click-able widget blocks and their respective _Parameters_ and _Input/Output_ tiles. 

```python
from chemml.wrapper.notebook import ChemMLNotebook
ui = ChemMLNotebook()
```

The workflow gives a precise representation of all the intermediate steps, blocks used to develop the model, the saved data and the inputs/outputs to each block. Once the workflow is finalized, we save the input script with our desired file name in _.txt_ format.

_Note:_ In this case, we specify our output directory as ‘read_excel’.

```python
from chemml.wrapper.engine import run
run(INPUT_FILE = '/mnt/c/Users/Aatish Pradhan/Desktop/master_chemml_wrapper/chemml/docs/read_excel.txt', OUTPUT_DIRECTORY = 'read_excel')
```
